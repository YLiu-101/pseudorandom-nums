```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
import math

```

```python
def c(num):
    return math.atan(num.imag/num.real) # the angle of the complex number
    
```

```python
z = complex(3,4)
z**z
total_deg = 900
arr = [c(a**2) for a in [complex(math.cos(2*math.pi*i/total_deg),math.sin(2*math.pi*i/total_deg)) for i in range(total_deg)]]

plt.plot(range(total_deg),2*np.array(arr)/math.pi)
```

```python
z = complex(3,2)
ar = [z]
for j in range(100):
    print(ar[-1])
    new_z = ar[-1]**z
    ar.append(new_z/abs(new_z))
plt.plot(range(101),[c(a) for a in ar])
```

```python
new_ar = [a for a in ar if a.imag>0]

# for i in range(len(ar)):
plt.plot(range(len(new_ar)),[c(a) for a in new_ar])
print(len(new_ar))
len(ar)

```
