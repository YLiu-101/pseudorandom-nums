```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
import math

```

```python
def c(num):
    return math.atan(num.imag/num.real) # the angle of the complex number
    
```

```python
z = complex(3,4)
z**z
total_deg = 900
arr = [c(a**2) for a in [complex(math.cos(2*math.pi*i/total_deg),math.sin(2*math.pi*i/total_deg)) for i in range(total_deg)]]

plt.plot(range(total_deg),2*np.array(arr)/math.pi)
```

```python
z = complex(3,0.5)/abs(z)
ar = [z]
n = 1000000
for j in range(n):
    # print(ar[-1])
    new_z = ar[-1]**z
    ar.append(new_z/abs(new_z))
plt.plot(range(n+1),[c(a) for a in ar])
```

```python
new_ar = [c(a) for a in ar if a.imag>0]

# for i in range(len(ar)):
plt.plot(range(len(new_ar)),new_ar)
print(len(new_ar))
len(ar)

```

```python

final_arr = [c(a)>0 for a in new_ar]
```

```python
print(sum(final_arr)/len(new_ar))
counter = 0
counter2 = 0
counter3 = 0
counter4 = 0
for i in range(1,len(final_arr)):
    if final_arr[i] == final_arr[i-1] and final_arr[i] == True:
        counter+=1
    if final_arr[i] == final_arr[i-1] and final_arr[i] == False:
        counter2+=1
    if final_arr[i] != final_arr[i-1] and final_arr[i] == True:
        counter3+=1
    if final_arr[i] != final_arr[i-1] and final_arr[i] == False:
        counter4+=1
print(counter, counter2, counter3, counter4) # For z=3+2i, T->T and F->F is more likely than T->F or F->T
```
